LINSPACE -- GENERATE NUMBERS BETWEN -10 TO +10

```python
# Topic - Array creation linspace -- ( GRAPH )Evenly spaced between two numbers
arr= np.linspace(0,1,5) # 0 to 1
print(arr)  # 0 0.25 0.5 0.75 1.

arr = np.linspace(0,5,10) # need 10 evenly spaced
print(arr)

arr=np.linspace(-10,10,5,dtype=int) # integer type data
print(arr)
```

```python
# Logspace --

arr= np.logspace(1,3,3)  # logartihmic scale array 10^1 10^2 10^3 .. 3 points
print(arr)

arr=np.logspace(1,3,2) # 2 points
print(arr)   # 10 1000
```

Array operations

```python
a1=np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(3,4)
print(a1)
print(a2)
```

```python
# scalar operations
# arithmethic
a=np.array([1,2,3,4,5])
print(a)
print(a+1)
print(a*2)
print(a/3)
print(a//2)

# relational operator - return boolean values
print(a>=5)
print(a<2)
print(a==2)


```

```python
# vector operations
print(a1)
print(a2)
print(a1+a2)
print(a1-a2)

print(a1>a2)
```

```python
# Array functions
a1=np.arange(1,11)
a1=np.random.random((3,3))
a1=np.round(a1*100)
print(a1)

# max min sum prod
print(np.max(a1)) # float value return
print(np.min(a1))
print(np.sum(a1))
print(np.prod(a1))


```

```python
# axis wise operation
print(a1)
print(np.max(a1,axis=0)) # max item in every column return
print(np.min(a1,axis=1))
print(np.sum(a1,axis=1))
print(np.prod(a1,axis=1))
```

```python
# mean median variance std
print(np.mean(a1))
print(np.mean(a1,axis=1))
print(np.std(a1,axis=1))
print(np.var(a1,axis=1))
print(np.median(a1))
print(np.median(a1,axis=1))
```

```python
# trignometric functions
print(np.sin(a1))
print(np.cos(a1))
print(np.tan(a1))
```

```python
# dot product  conditon
print(a1)
print(a2)
print(np.dot(a1,a2))
```

```python
# log and exp
print(np.log(a1))
print(np.exp(a1))
```

```python
# round floor ceil
a=np.random.random((2,3))*100
print(a)
print(np.round(a))
print(np.floor(a))
print(np.ceil(a))
```

```python
a1=np.arange(20)
a2=np.arange(12).reshape(3,4)
a3=np.arange(8).reshape(2,2,2)

# slicing and indexing
print(a1)
# acess using indexing
print(a1[0],a1[10]) # one and two also

```

SLICING(imp)

```python
# SLICING IN 1D
print(a1)


# slicing
print(a1[1:6])

# 10 to 12 only

print(a1[10:13])
```

```python
# slicing in 2d
print(a2)

# 1st row
print(a2[:1])

# third column
print(a2[:,2])
print(a2[2:]) # 8 9 10 11

# 5 6
# 9 10
print(a2[1:,1:3])


# 0 3
# 8 11
print(a2[::2,::3]) # excusive so 3
```
