```python
# SORTING (sort) np.sort
# Return a sorted copy of an array
import numpy as np
a=np.random.randint(1,100,15)
b=np.random.randint(1,100,24).reshape(6,4)
print(a)
print(b)

# sort
print(np.sort(a)) # aascending order sorting

```

```python
# descending order sorting
print(np.sort(a)[::-1])
```

```python
# row wise and column wise sorting using axis
print(b)
np.sort(b,axis=0)
np.sort(b,axis=1) # column wise
```

```python
# 2- function np.append
print(a)
print(np.append(a,200)) # append add on the last


# for 2d array
print(np.append(b,np.ones((b.shape[0],1)),axis=1))  # row

# column
print(np.append(b,np.random.random((b.shape[0],1)),axis=1))
```
