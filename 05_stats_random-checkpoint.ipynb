```python
import numpy as np
arr = np.random.rand(10) # random floats (0 t0 1 )  uniformly distributed generate value 0 to 1
print(arr)

# generate 10 value random from 0 to 1
# 2
arr=np.random.rand(3, 3) # random value  0 to 1 total elements 9 
print(arr)
```

```python
# 2--( Normal distribution )  -- random number having a normal distribution
# noth positive and negative 
arr = np.random.randn(10) 
print(arr)
```

```python
arr = np.random.randn(2,3)
print(arr)
```

```python
# 3 random integers ( start stop dimension size value)
arr = np.random.randint(10)
print(arr)

arr = np.random.randint(10,100) 
print(arr) 

arr = np.random.randint(2,10,5) # here 5 random value between 2 to 10 
print(arr)
```
