```python
# Array Creation #--2
import numpy as np
a1=np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(3,4)
print(a1)
print(a2)
```

```python
# scalar operations / arithmethic
print(a1*2)
print(a1+2)
print(a1-2)
print(a1%2)
print(a1/2)
```

```python
# Relational operations
print(a1)
print(a1>5)
print(a1==5)
print(a1!=5)
```

```python
# Vector operations arithmetic  "3d item wise add first item of a1 + first item of a2 as it soon "
print(a1)
print(a2)
print(a1+a2)
```

```python
# Array functions
a1=np.arange(1,11)
a2=np.random.random((3,3))  # float value  return by default
print(a1)
print(a2)
round=np.round(a2)
print(round)
print(np.sum(a1))

```

```python
print(np.max(a1))
print(np.min(a2))
print(np.sum(a1))
print(np.prod(a1))

```

```python
# for particular row or column perform operation
a=np.arange(1,10).reshape(3,3)
print(a)
print(np.max(a,axis=0)) # 7 8 9
print(np.min(a,axis=1))

# sum and product
print(np.sum(a,axis=1))
print(np.prod(a,axis=0))
```

```python
# mean median std var
print(np.mean(a))
print(np.median(a))
print(np.std(a))
print(np.var(a))
```

```python
# mean medain var standard deviation range axis wise
print(np.mean(a,axis=1))
print(np.median(a,axis=1))
print(np.var(a,axis=1))
print(np.std(a,axis=1))
```
